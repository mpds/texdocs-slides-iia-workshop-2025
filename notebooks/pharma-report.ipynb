```python
import matplotlib.pyplot as plt
import scienceplots

plt.style.use(['science', 'nature', 'muted'])  # 'nature', 'ieee', 'science', 'grid', 'no-latex'
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

# colour palette
```

```python
# pipeline sizes
n_cmpds = [12704, 3954, 3586, 1358, 294, 189]

# development phases
phases = ["Pré-clínico", "Fase I", "Fase II", "Fase III", "Pré-registro", "Registro"]
```

```python
# plot
fig, ax = plt.subplots(dpi=300)
bars = ax.bar(phases, n_cmpds)

# Add numbers on top of each bar
for bar, value in zip(bars, n_cmpds):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
            f'{value}',
            ha='center', va='bottom')

# ax.set_yscale('log')
ax.set_ylabel('Quantidade de compostos')
# ax.set_title('Número de Compostos por Fase de Desenvolvimento')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# change yaxis to scientific notation
ax.ticklabel_format(axis='y', style='sci', scilimits=(0,0))

# change yaxis range
ax.set_ylim(0, 14000)

# save pdf
plt.savefig('pipeline_sizes.pdf')
plt.show()
```
